```python
# 计算每个公司的记录数
company_counts = df.groupby("Stkcd").Accper.count()
# 筛选出记录数大于等于3的公司
valid_companies = company_counts[company_counts >= 3].index
# 只保留这些公司的数据
df = df[df.Stkcd.isin(valid_companies)]
# 显示筛选后每个公司的记录数
df.groupby("Stkcd").Accper.count()
```

```python
# 首先按公司和年份排序
df = df.sort_values(by=["Stkcd", "Accper"])

# 计算每个公司的连续年份
def has_consecutive_years(group):
    years = group["Accper"].values
    if len(years) < 3:
        return False
    # 检查年份是否连续
    for i in range(len(years) - 2):
        if years[i + 2] - years[i] == 2:  # 如果当前年份和第2个年份之间相差2年，那么是连续的
            return True
    return False

# 按公司分组，筛选出至少有3年连续年份的数据
valid_companies = df.groupby("Stkcd").filter(has_consecutive_years)

# 查看筛选后的结果
valid_companies.groupby("Stkcd").Accper.count()
```

```python
valid_companies.to_csv("valid_companies_with_more.csv", index=False)
```

```python
# 筛选出记录数为14的公司
company_counts = valid_companies.groupby("Stkcd").Accper.count()
companies_with_14_records = company_counts[company_counts == 14]
print("记录数为14的公司：")
print(companies_with_14_records)
```

```python
# 筛选出记录数为14的公司的数据
companies_with_14_records = company_counts[company_counts == 14].index
filtered_df = valid_companies[valid_companies.Stkcd.isin(companies_with_14_records)]

# 查看筛选结果
print(f"筛选出的公司数量: {len(companies_with_14_records)}")
print(f"筛选出的记录总数: {len(filtered_df)}")

# 显示这些公司的数据
print("\n筛选出的公司数据预览:")
display(filtered_df.head())

# 按公司分组并确认每个公司都有14条记录
verification = filtered_df.groupby("Stkcd").Accper.count()
print("\n确认每个公司的记录数:")
display(verification)
```

```python
# 按年份聚合并对每个年份内的财务指标进行归一化
# 首先确定需要归一化的列（排除非数值列和标识列）
numeric_columns = filtered_df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print(len(numeric_columns))
# 排除Stkcd, Accper, STPT等非财务指标列
columns_to_normalize = [col for col in numeric_columns if col not in ['Stkcd', 'Accper', 'STPT']]

# 创建一个新的DataFrame来存储归一化后的数据
filtered_df = filtered_df.copy()

# 按年份分组并对每个财务指标进行归一化
for year, year_group in filtered_df.groupby('Accper'):
    # 对每个年份内的财务指标进行归一化
    for col in columns_to_normalize:
        # 计算该年份该指标的均值和标准差
        mean_val = year_group[col].mean()
        std_val = year_group[col].std()
        
        # 避免除以0
        if std_val == 0:
            std_val = 1
            
        # 对该年份的该指标进行归一化
        filtered_df.loc[year_group.index, col] = (year_group[col] - mean_val) / std_val

print(f"归一化前的数据示例:\n{filtered_df[['Accper'] + columns_to_normalize[:5]].head()}")
print(f"\n归一化后的数据示例:\n{filtered_df[['Accper'] + columns_to_normalize[:5]].head()}")

# 显示归一化后的完整数据集
filtered_df
```
